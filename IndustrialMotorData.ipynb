# Library import

```python
import pandas as pd
import sqlite3
```

# Load data from dataset

```python
df = pd.read_csv('data.csv')
print(df.head())
print(df.columns)
```

```python
# Clear the dataset with the columns we want to keep
columns = ['timestamp', 'machine_id', 'temperature', 'vibration',
       'pressure', 'energy_consumption', 'machine_status', 'anomaly_flag']

df_filtered = df[columns]

# Filter for just 10 random machines, instead of the 50 in the original dataset
unique_machines = df_filtered["machine_id"].drop_duplicates().sample(10, random_state=42)
df_10 = df_filtered[df_filtered["machine_id"].isin(unique_machines)]

# Save the cleaned dataset to a new CSV file
df_10.to_csv('cleaned_data.csv', index=False)
```

# Connect to Database and load the DB

```python
# Create SQL connection
conn = sqlite3.connect('industrial_data.db')

# Create a new table and insert the cleaned data into the SQL database
df_10.to_sql('motor_data', conn, if_exists='replace', index=False)

# Query the database to check the inserted data
query = "SELECT * FROM motor_data"
df_10 = pd.read_sql_query(query, conn)
print(df_10)

# Close the SQL connection
conn.close()
```
